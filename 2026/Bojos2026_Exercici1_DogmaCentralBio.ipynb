¡Bienvenidos! En este ejercicio empezaremos a programar para entender uno de los procesos fundamentales de la naturaleza: cómo la información contenida en nuestro DNA se utiliza para formar **proteínas**, las moléculas que se encargan de llevar a cabo funciones celulares de todo tipo.

El "Dogma Central de la Biología Molecular" explica el flujo de la información genética dentro de una célula. Este proceso tiene dos pasos principales:

1. **Transcripción (DNA → RNA):** Si el DNA es el "manual de instrucciones", el RNA sería una "fotocopia" temporal de una página del manual, aquélla que indica fabricar un objeto en concreto que necesitamos en ese momento.

2. **Traducción (RNA → Proteína):** El proceso de traducción sería análogo a leer esa fotocopia para ensamblar las piezas y construir un objeto (una proteína funcional).

 Por tanto, partir del fragmento de DNA que interese, se forma un RNA mensajero (mRNA). A continuación, unas moléculas llamadas **ribosomas** se unen al mRNA y lo traducen a la proteína correspondiente. Una vez formadas, las proteínas se encargan de llevar a cabo todo tipo de funciones celulares.


#### **Conceptos clave iniciales**

* *El "alfabeto" del DNA:* El DNA está formado por 4 letras (nucleótidos): **A** (Adenina), **T** (Timina), **C** (Citosina) y **G** (Guanina). El DNA está formado por dos cadenas que se unen como una cremallera siguiendo una regla estricta: la **A siempre interacciona con la T**, y la **C siempre con la G**.

* *El "alfabeto" del RNA:* Es casi idéntico al del DNA, pero la Timina (T) se sustituye por el **Uracilo (U)**.

* *El "alfabeto" de las proteínas:* Las proteínas son secuencias de pequeñas moléculas llamadas **aminoácidos (aa)**. En realidad, una secuencia de aa recibe el nombre de **péptido**. Cuando los péptidos se pliegan y adquieren estructuras tridimensionales, que les permiten realizar funciones concretas, pasamos a hablar de proteínas.

## **1.1.TRANSCRIPCIÓN: DEL DNA AL mRNA**

En las células eucariotas, el DNA está guardado de forma segura en el núcleo de la célula. Para llevar las instrucciones a las "fábricas" de proteínas (los ribosomas) que están fuera del núcleo, la célula crea una copia llamada **RNA mensajero (mRNA)**.

En bioinformática, simular esto es muy sencillo: solo tenemos que cambiar todas las letras 'T' de nuestra secuencia de ADN por letras 'U'.

Supongamos que tengamos una  secuencia de DNA (```dna_seq```) y queremos investigar qué proteína se formará a partir de un gen aquí contenido. El primer paso será predecir qué mRNA se formará a partir de este fragmento de DNA.






````python
# Definimos nuestra secuencia de ADN inicial (una cadena de texto)
dna_seq = "CGTATGGCGGACCCAGAATGAGGATAAGACCGTTAGATACGC"

print(f"Nuestra secuencia de DNA es: {dna_seq}")
````

Utilizando Python, creamos la función ``` transcribe() ```. Esta función toma como input cualquier secuencia de DNA, y devuelve como output el correspondiente mRNA.

````python
def transcribe(secuencia_dna):
    """
    Convierte una secuencia de DNA a RNA reemplazando la Timina (T) por Uracilo (U).
    """
    # Usamos la función .replace() de Python
    mRNA = secuencia_dna.replace('T', 'U')

    return mRNA
````

Una vez definida, podemos aplicar la función ```transcribe()``` a nuestra secuencia de DNA para simular la transcripción y obtener las secuncia del mRNA que se formará.

````python
# Aplicamos nuestra función a la secuencia de DNA para obtener
# el correspondiente mRNA
rna_seq = transcribe(dna_seq)

print(f"DNA:  {dna_seq}")
print(f"mRNA: {rna_seq}")
````

## **1.2.TRADUCCIÓN: DEL mRNA A PROTEÍNA**

El mRNA es sólo un paso intermedio entre el DNA y el objetivo final: "construir" una proteína. La maquinaria de transcripción identifica, dentro de todo el DNA, qué parte contiene la información necesaria para formar una proteína determinada, y extrae esa información en forma de mRNA. Una vez el mRNA sale del núcleo, los ribosomas lo leen para construir la proteína correspondinte.

Los ribosomas tienen una manera muy concreta de fabricar las proteínas. No leen letra por letra, sino en **grupos de 3 letras** que llamamos **codones**. Cada codón codifica un aminoácido distinto. Es decir, el ribosoma mira 3 letras a la vez y, dependiendo de qué letras sean, coloca un aminoácido u otro. Para saber qué pieza (aminoácido) corresponde a cada codón, la célula usa el **Código Genético**, que es universal para casi todos los seres vivos.

De este modo, si, por ejemplo, el ribosoma ve el codón "CCA", colocará una leucina (leu). Si a continuación aparece el codón "AGC", seguirá una serina (ser), etc.


**Importante!:** Los ribosomas siguen unas "instrucciones" para saber dónde empezar a formar una proteína, y dónde acabar. Hay unos codones específicos que marcan siempre el inicio de una proteína y el final de una proteína.


* **Codón de INICIO (START):** `AUG`. Le dice al ribosoma: *"¡Empieza a construir la proteína aquí!"*. Además, codifica para el aminoácido Metionina (M).

* **Codones de PARADA (STOP):** `UAA`, `UAG`, `UGA`. Le dicen al ribosoma: *"¡Para! La proteína está terminada"*. No añaden ningún aminoácido (los representaremos con un `*`).

#### **Concepto importante: Open Reading Frames (ORFs)**

Dado que leemos el mRNA de 3 en 3, importa muchísimo **dónde empezamos a leer**. Según cómo agrupemos los aminoácios, se formarán codones diferentes, y por consiguiente se formarán proteínas distintas.

Por ejemplo, imaginemos que tenemos el siguiente mRNA: "CAGGUGAUGUUUCCAUCAUAGACA". Hay tres posibles ORFs (Open Reading Frames/Marcos de lectura):

* *ORF1 (Empezando en la posición 0)*: CAG-GUG-AUG-UUU-CCA-UCA-UAG-ACA

* *ORF2 (Empezando en la posición 1)*: *C*- AGG-UGA-UGU-UUC-CAU-CAU-AGA-*CA*

* *ORF3 (Empezando en la posición 2)*: *CA*-GGU-GAU-GUU-UCC-AUC-AUA-GAC-*A*

En teoría, podría ocurrir que todos los ORFs de un mRNA pudieran ser traducidos a proteínas. En este caso, cada ORF
 daría lugar a una proteína distinta (esto es algo muy común en virus, que tienen que "optimizar" el espacio en sus genomas). Sin embargo, la realidad es que no todos los ORFs son válidos: **Un ORF válido debe tener un codón de inicio (`AUG`) y terminar con un codón STOP.**

Al igual que con la transcripción, podemos crear una función para la traducción (`translate()`), que tome como input un mRNA y prediga las proteínas que se pueden formar tomándolo como molde, teniendo en cuenta todos los posibles ORFs.

Para implementarla, el primer paso consiste definir una variable, `codon_table`  (expresado con un "diccionario" en Python), que contenga el código genético. En este diccionario, a cada codón de 3 letras (clave) le asigna una letra de 1 aminoácido (valor). Tal y como funcionan los diccionarios en Python, teniendo un codón como clave, podremos usar `codon_table`para saber qué aa le corresponde.

Por ejemplo, si, leyendo un mRNA, encontramos el codón "AGC", buscaremos "AGC" en el diccionario (clave), y encontraremos que corresponde con el valor "S", correspondiente al aminoácido "serina".

````python
codon_table = {
    'AUA':'I', 'AUC':'I', 'AUU':'I', 'AUG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACU':'T',
    'AAC':'N', 'AAU':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGU':'S', 'AGA':'R', 'AGG':'R',
    'CUA':'L', 'CUC':'L', 'CUG':'L', 'CUU':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCU':'P',
    'CAC':'H', 'CAU':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGU':'R',
    'GUA':'V', 'GUC':'V', 'GUG':'V', 'GUU':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCU':'A',
    'GAC':'D', 'GAU':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGU':'G',
    'UCA':'S', 'UCC':'S', 'UCG':'S', 'UCU':'S',
    'UUC':'F', 'UUU':'F', 'UUA':'L', 'UUG':'L',
    'UAC':'Y', 'UAU':'Y', 'UAA':'*', 'UAG':'*', 'UGA':'*' # * = Codón STOP
}
````

Construimos la función `translate()`, que toma como input una secuencia de mRNA y el ORF elegido, y traduce el mRNA simulando el comportamiento real de un ribosoma. Es muy importante asegurarnos de que el mRNA sólo se traduzca si hay un codón de inicio y un codón STOP en el mismo ORF.

````python
def translate(secuencia_rna, frame=0): # En los argumentos de la función,
    # indicamos que, por defecto, el ORF será el primero (empezando a leer)
    # en la primera letra. Sin embargo, cuando utilicemos la función, podemos
    # cambiarlo a cualquiera de los otros dos ORFs
    """
    Traduce el RNA a proteína simulando el comportamiento real de un ribosoma.
    Solo empieza al encontrar 'AUG' y se detiene al encontrar un STOP.
    """
    proteina = ""
    traduccion_iniciada = False

    # Recorremos el ARN de 3 en 3 letras, empezando desde el 'marco' (ORF) elegido (0, 1 o 2)
    for i in range(frame, len(secuencia_rna) - 2, 3):
        codon = secuencia_rna[i:i+3]
        aminoacido = codon_table.get(codon, '?')

        # REGLA 1: Buscar el codón de INICIO ('AUG')
        if not traduccion_iniciada:
            if codon == 'AUG':
                traduccion_iniciada = True
                proteina += aminoacido

            continue # El ribosoma ignora todo hasta encontrar el inicio

        # REGLA 2: Si ya empezamos, seguimos añadiendo aminoácidos
        if traduccion_iniciada:
            # REGLA 3: Parar si encontramos un STOP ('*')
            if aminoacido == '*':
                return f"✅ Éxito: Proteína válida encontrada en el ORF {frame} -> {proteina}*"

            proteina += aminoacido

    # REGLA 4: Si se acaba el RNA y llegamos aquí, algo ha fallado.
    if not traduccion_iniciada:
        return f"❌ Error en el ORF {frame}: No se encontró ningún codón de INICIO (AUG)."
    else:
        return f"⚠️ Error en el ORF {frame}: Se empezó a traducir, pero el mRNA terminó sin un codón de STOP. Proteína parcial: {proteina}"
````

¡Vamos a poner a prueba nuestra función! Vamos a leer nuestro mRNA usando los tres ORFs posibles para investigar si alguno de ellos da lugar a una proteína completa.

````python
# ORF1 (Empezando a leer desde la primera letra)
print(translate(rna_seq, frame=0))
````

````python
# ORF2 (Nos saltamos la primera letra y empezamos a leer de 3 en 3 desde la segunda)
print(translate(rna_seq, frame=1))
````

````python
# ORF3 (Nos saltamos las dos primeras letra y empezamos a leer de 3 en 3 desde la tercera)
print(translate(rna_seq, frame=2))
````
